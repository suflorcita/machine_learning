```python
import pandas as pd
import matplotlib.pyplot as plt 
```

La idea es: 
- Convertir las fechas a un formato de mes (están en MM/DD/YYYY)
- Extraer el mes a una nueva columna



```python

pumpkins = pd.read_csv('US-pumpkins.csv')

pumpkins.head() # ya a simple vista se ven los NaN
```

```python
pumpkins.tail(5) # últimas 5 lineas
```

```python
pumpkins.isnull().sum() #cuenta los NaN en cada columna 
```

Pero tengo un problema... las calabazas se venden con diferentes unidades de medidas. Por eso vamos a filtrar solo las que tienen la palabra "bushel" en la columna Package.

```python
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
```

```python
pumpkins # solo me quedan 415 filas 
```

```python
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date'] # son las unicas columnas que me interesan
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1) 
# bajo las columnas que no me interesan
```

```python
pumpkins
```

Quiero un promedio del precio en cada mes: uso un promedio entre el precio más alto y el más bajo

```python
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2 # creo una nueva columna
```

```python
price
```

```python
month = pd.DatetimeIndex(pumpkins['Date']).month # separo el mes de la fecha
```

```python
month
```

```python
print(month)
```

Hay un problema, el precio no está normalizado por unidad. Algunas son de 1/2 bushel, otras de 1 1/9 bushel, vamos a crear un nuevo data frame limpio y normalizar: 

```python
new_pumpkins = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'],'High Price': pumpkins['High Price'], 'Price': price})
# hago un dataframe "limpio "
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)

new_pumpkins
```

Es facilmente observable que las calabazas que se venden por 1/2 bushel son mucho más caras...

### Estrategias de Visualización 

```python
price = new_pumpkins.Price
month = new_pumpkins.Month
plt.scatter(price, month) #Scatter plot. 
plt.show()
```

Este gráfico no es muy útil

```python
new_pumpkins.groupby(['Month'])['Price'].mean().plot(kind='bar')
plt.ylabel("Pumpkin Price")
```

Hace un gráfico de barra en el que se grafica el precio promedio de la calabaza por mes. Las ventas aumentan en octubre -> por Halloween
