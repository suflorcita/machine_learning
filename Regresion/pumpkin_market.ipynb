La idea es: 
- Convertir las fechas a un formato de mes (están en MM/DD/YYYY)
- Extraer el mes a una nueva columna

```python
import pandas as pd
pumpkins = pd.read_csv('US-pumpkins.csv')
pumpkins.head() # ya a simple vista se ven los NaN
```

```python
pumpkins.tail(5) # últimas 5 lineas
```

```python
pumpkins.isnull().sum() #cuenta los NaN en cada columna b
```

```python
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date'] # son las unicas columnas que me interesan
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1) 
# bajo las columnas que no me interesan
```

```python
pumpkins
```

Quiero un promedio del precio en cada mes: uso un promedio entre el precio más alto y el más bajo

```python
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2 # creo una nueva columna
```
